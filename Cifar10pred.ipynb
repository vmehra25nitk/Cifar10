```python
import cifar10
import matplotlib.pyplot as plt
cifar10.data_path = "data/CIFAR-10/"
class_names = cifar10.load_class_names()
images_train, cls_train, labels_train = cifar10.load_training_data()
images_test, cls_test, labels_test = cifar10.load_test_data()
```

```python
from sklearn.decomposition import PCA
import numpy as np
```

Flatten images into a 1d array

```python
x_train = images_train.reshape(images_train.shape[0], -1)
x_test = images_test.reshape(images_test.shape[0], -1)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
y_train = cls_train
y_test = cls_test
```

```python
pca = PCA()
pca.fit(x_train)
```

```python
k = 0
total = sum(pca.explained_variance_)
current_sum = 0
while (current_sum / total) < 0.99:
    current_sum += pca.explained_variance_[k]
    k += 1
print(k)
```

```python
pca = PCA(n_components = k, whiten = True)
```

```python
x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn import svm
from sklearn.svm import SVC
from sklearn import tree
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
import pickle
pickle_in = open("x_train_pca.txt", 'rb')
x_train_pca = pickle.load(pickle_in)
pickle_in = open("x_test_pca.txt", 'rb')
x_test_pca = pickle.load(pickle_in)

pickle_out = open('x_train_pca.txt', 'wb')
pickle.dump(x_train_pca, pickle_out)
pickle_out.close()

pickle_out = open('x_test_pca.txt', 'wb')
pickle.dump(x_test_pca, pickle_out)
pickle_out.close()
```

```python
print(x_train_pca.shape)
print(x_test_pca.shape)
```

```python
svm_clf = svm.SVC()
svm_clf.fit(x_train_pca, y_train)
x_test_pca = pca.transform(x_test)
```

```python
knn_clf = KNeighborsClassifier()
knn_clf.fit(x_train_pca, y_train)
```

```python
nb_clf = GaussianNB()
nb_clf.fit(x_train_pca, y_train)
```

```python
rdm_clf = RandomForestClassifier()
rdm_clf.fit(x_train_pca, y_train)
```

```python
y_pred_rdm = rdm_clf.predict(x_test_pca)
y_pred_nb = nb_clf.predict(x_test_pca)
y_pred_knn = knn_clf.predict(x_test_pca)
```

```python
print(accuracy_score(y_pred_knn, y_test))
```
