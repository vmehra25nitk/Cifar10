```python
import cifar10
import matplotlib.pyplot as plt
cifar10.data_path = "data/CIFAR-10/"
class_names = cifar10.load_class_names()
images_train, cls_train, labels_train = cifar10.load_training_data()
images_test, cls_test, labels_test = cifar10.load_test_data()
```

```python
from sklearn.decomposition import PCA
import numpy as np
```

```python
fig = plt.figure(figsize = (8, 8))
for i in range(64):
    ax = fig.add_subplot(8, 8, i+1)
    ax.imshow(images_train[i], cmap = plt.cm.bone)
plt.show()
```

Flatten images into a 1d array

```python
x_train = images_train.reshape(images_train.shape[0], -1)
x_test = images_test.reshape(images_test.shape[0], -1)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
y_train = cls_train
y_test = cls_test
```

```python
pca = PCA()
pca.fit(x_train)
```

```python
k = 0
total = sum(pca.explained_variance_)
current_sum = 0
while (current_sum / total) < 0.99:
    current_sum += pca.explained_variance_[k]
    k += 1
print(k)
```

```python
pca = PCA(n_components = k, whiten = True)
```

```python
x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn import svm
from sklearn.svm import SVC
from sklearn import tree
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
import pickle
pickle_in = open("x_train_pca.txt", 'rb')
x_train_pca = pickle.load(pickle_in)
pickle_in = open("x_test_pca.txt", 'rb')
x_test_pca = pickle.load(pickle_in)
```

```python
rdm_clf = RandomForestClassifier()
rdm_clf.fit(x_train_pca, y_train)
```

```python
def convertToClass(array_num, class_names):
    result = []
    for i in array_num:
        i = int(i)
        result.append(class_names[i-1])
    return result
```

```python
y_pred_rdm = np.loadtxt('Ansrdm.csv')
y_pred_nb = np.loadtxt('Ansnb.csv')
y_pred_rdm = convertToClass(y_pred_rdm, class_names)
y_pred_nb = convertToClass(y_pred_nb, class_names)
```

```python
np.savetxt('Ansrdm.csv', y_pred_rdm, delimiter = ",", fmt = "%s")
np.savetxt('Ansnb.csv', y_pred_nb, delimiter = ",", fmt = "%s")
#np.savetxt('Ansknn.csv', y_pred_knn, fmt = '%.0f', delimiter = ',')
```
